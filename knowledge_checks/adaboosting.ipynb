```python
# Import Python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create Data
dat = pd.DataFrame({'x': [0,1,2],
                    'y': [-1,1,-1],
                    'c': ['red', 'blue', 'red']})

# Scatter Plot
plt.scatter(x = dat['x'], y = dat['y'], c=dat['c'])
```

```python
# Question 2a
def h1_x(x):
    result = x - 0.5
    if result < 0:
        return(-1)
    else:
        return(1)

line_x = np.linspace(0,2,100)
line_y = [h1_x(x) for x in line_x]

# Plot
plt.scatter(x = dat['x'], y = dat['y'], c=dat['c'])
plt.plot(line_x, line_y, 'k--')
```

```python
# Question 2b - renormalize weights after first round result
error = 1/3 # Third point misclassified
c = np.log((1-error)/error)
weights = [1/3, 1/3, (1/3)*np.exp(c)]
normalized_weights = weights/np.sum(weights)
iterator = 1
print('Weights after first round: ')
for error in normalized_weights:
    print('w'+str(iterator)+': ' + str(error))
    iterator+=1
```

```python
# Question 3a
def h2_x(x):
    result = x - 1.5
    if result < 0:
        return(1)
    else:
        return(-1)

line_x = np.linspace(0,2,100)
line_y = [h2_x(x) for x in line_x]

# Plot
plt.scatter(x = dat['x'], y = dat['y'], c=dat['c'])
plt.plot(line_x, line_y, 'k--')
```
